# Matriz de covarianza

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')
iris.dtypes
```

```python
# Visualizar la matriz de covarianza
sns.pairplot(iris)
```

```python
# Categorizar la matriz de covarianza usando el argumento "hue"
sns.pairplot(iris, hue='species')
```

```python
# Visualizar una lista de las columnas del dataset para luego aplicar el siguiente paso
iris.columns
```

El siguiente script corresponde matemáticamente a aplicar la **normalización estándar (z
-score normalizada)** a restar por el promedio y dividir por la desviación estándar por cada variable. Es decir:
* StandarScaler: ajusta los datos para que cada columna tenga una media de 0 y una desviación estándar de 1, de tal forma que, se pueda comparar las columnas sin que influya por las escalas diferentes que puedan tener.
* Transormación de los datos: aquí se aplica la fórmula donde a cada valor se le resta la media de los valores de la columna y se divide luego por la desviación estándar z = (x - μ)/σ. Esto da como resultado el valor escalado o transformado.

`fit_transform` calcula:
* La media (μ) y la desviación estándar (σ) para cada columna.
* Luego, aplica la fórmula para calcular (z) a todos los valores de las columnas.
* La variable `scaled` será entonces un arreglo NumPy con los valores escalados.

Finalmente, las columnas transformadas tendrán:
* Una media cercana a 0
* Una desviación estándar de 1

```python
# Crear el escalamiento de los datos
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled
```

```python
# Aplicar la transpuesta para poder usar la matriz de correlación
# Cada arreglo [] serán los datos de cada variable
scaled.T
```

```python
# Calcular la matriz de covarianza
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
# Graficar la matriz de covarianza en un mapa de calor
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix,
                cmap="RdBu",
                cbar=True,  # Escala de colores
                annot=True, # Mostrar valores en cada elemento.
                square=True,    # Matriz la muestre cuadrada
                fmt='.2f',
                annot_kws={'size':12},
                yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
                )
```
